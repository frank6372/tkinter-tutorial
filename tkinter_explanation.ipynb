```python
from tkinter import *

master = Tk()
var1 = IntVar()
Checkbutton(master, text='male', variable=var1).grid(row=0, sticky=W)
var2 = IntVar()
Checkbutton(master, text='female', variable=var2).grid(row=1, sticky=W)
mainloop()
```

In your code, `IntVar` is a class from the `tkinter` module used to create an integer variable that can be shared between different widgets. Here's a breakdown of how it's used in your example:

- `var1 = IntVar()` creates an integer variable `var1`.
- `Checkbutton(master, text='male', variable=var1)` creates a checkbutton labeled "male" and links it to `var1`.
- `var2 = IntVar()` creates another integer variable `var2`.
- `Checkbutton(master, text='female', variable=var2)` creates a checkbutton labeled "female" and links it to `var2`.

When you check or uncheck these buttons, the values of `var1` and `var2` will change accordingly. This allows you to track the state of each checkbutton (checked or unchecked) through these variables.


The `IntVar` class is used to create an integer variable in `tkinter`. It's specifically designed to hold integer values, which is useful for widgets like checkbuttons, radiobuttons, and scales that typically work with integer states (e.g., checked/unchecked, selected/unselected).

Using `IntVar` ensures that the variable associated with the widget will always be an integer, which simplifies handling the widget's state. For example, a checkbutton will set the `IntVar` to `1` when checked and `0` when unchecked.

If you need to store other types of data, `tkinter` also provides `StringVar`, `DoubleVar`, and `BooleanVar` for string, floating-point, and boolean values, respectively.



#### Radio button
```python
from tkinter import *

root = Tk()
v = IntVar()
Radiobutton(root, text='GfG', variable=v, value=1).pack(anchor=W)
Radiobutton(root, text='MIT', variable=v, value=2).pack(anchor=W)
mainloop()
```
It is used to offer multi-choice option to the user. It offers several options to the user and the user has to choose one option.

In Tkinter, the `anchor` option is used to define where the widget (in this case, the `Radiobutton`) is positioned relative to its reference point. The `anchor=W` option means that the widget will be anchored to the **west** (left) side of its containing widget. This means the left edge of the `Radiobutton` will align with the left edge of its container².

Here's a quick example to illustrate:

- `anchor=W`: Aligns the widget to the left.
- `anchor=E`: Aligns the widget to the right.
- `anchor=CENTER`: Centers the widget.

#### Scroll bar
```python
from tkinter import *

root = Tk()
scrollbar = Scrollbar(root)
scrollbar.pack(side=RIGHT, fill=Y)
mylist = Listbox(root, yscrollcommand=scrollbar.set)

for line in range(100):
    mylist.insert(END, 'This is line number' + str(line))
    
mylist.pack(side=LEFT, fill=BOTH)
scrollbar.config(command=mylist.yview)
mainloop()
```

#### Menu

```python
from tkinter import *

root = Tk()
menu = Menu(root)
root.config(menu=menu)
filemenu = Menu(menu)
menu.add_cascade(label='File', menu=filemenu)
filemenu.add_command(label='New')
filemenu.add_command(label='Open...')
filemenu.add_separator()
filemenu.add_command(label='Exit', command=root.quit)
helpmenu = Menu(menu)
menu.add_cascade(label='Help', menu=helpmenu)
helpmenu.add_command(label='About')
mainloop()
```

#### Combobox

```python
import tkinter as tk
from tkinter import ttk

def on_select(event):
    selected_item = combo_box.get()
    label.config(text="Selected Item: " + selected_item)

root = tk.Tk()
root.title("Combobox Example")

# Create a label
label = tk.Label(root, text="Selected Item: ")
label.pack(pady=10)

# Create a Combobox widget
combo_box = ttk.Combobox(root, values=["Option 1", "Option 2", "Option 3"])
combo_box.pack(pady=5)

# Set default value
combo_box.set("Option 1")

# Bind event to selection
combo_box.bind("<<ComboboxSelected>>", on_select)

```
